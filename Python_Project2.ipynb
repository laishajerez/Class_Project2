```python
# Create a subset of the original dataframe: with Gender, age group, and readmitted numbers
df3=df[['age','readmitted']]
print(df3)
```

```python
# Group by 'age' and count the occurrences of each 'readmitted' value
grouped_df = df3.groupby(['age', 'readmitted']).size().reset_index(name='count')

print(grouped_df)
```

```python
import seaborn as sns
```

```python
# Pivot the DataFrame to have 'age' as rows, 'readmitted' as columns, and 'count' as values
pivot_df = grouped_df.pivot(index='age', columns='readmitted', values='count').fillna(0)

# Plot a bar chart
plt.figure(figsize=(10, 6))
pivot_df.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Readmitted Cases by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Readmitted')
plt.show()
```
